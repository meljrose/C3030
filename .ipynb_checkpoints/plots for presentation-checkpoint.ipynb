```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.cm as cm
from collections import OrderedDict
from matplotlib import rc
import matplotlib as mpl
import matplotlib.cm as cm
```

```python
rc('font',**{'family':'serif','serif':['serif'],'size':16})
```

```python
class Survey(object):
    def __init__(self, name, freq, year, sensitivity, angular_res, color):
        self.name = name
        self.freq = freq
        self.year = year
        self.angular_res = angular_res
        self.sensitivity = sensitivity
        self.color = color


        
# double check the scalings        
surveys = [Survey('VLSSr', [0.074], [2001, 2007], [100.], 80., 'DarkOrchid'), Survey('TGSS', [0.148], [2010,2012], [2,5], 25., 'black'), 
           Survey('MRC', [0.408], [1968,1978], [150.], 180., 'forestgreen'), Survey('SUMSS', [0.843], [1997,2003], [1.], 45.,'saddlebrown'), 
           Survey('NVSS', [1.40], [1993,1996], [0.45], 45., 'navy'), Survey('ATPMN', [4.85,8.64], [1992,1994], [1.], 1.2, 'darkgreen'),
           Survey('AT20G', [4.8, 8.64, 19.904], [2004,2008], [40,100], 10., 'MediumSeaGreen'),
           Survey('ATCA', [2.1,5.5,9.], [2015,2016], [0.00], 0.00, 'indigo'),Survey('GLEAM', [0.076, 0.227], [2013,2016], [6,10], 100.00, 'crimson')]


```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

colors = cm.rainbow(np.linspace(0, 1, len(surveys)))
n = 3
width = 1
y = np.linspace(0, 1, len(surveys))
for survey, color in zip(surveys, colors): 
    #s = 2*2**(survey.angular_res/10)#[20*2**n for n in range(len(x))]
    c = survey.sensitivity

    ax.fill_betweenx(np.linspace(y-width, y+width, n), np.repeat(survey.year[1], n), np.repeat(2017, n), 
                     facecolor = color, edgecolor='none', label = survey.name)
    ax.plot([], [], color=color,label = survey.name)
    #plt.fill_between(survey.freq, [survey.year[0]], [survey.year[1]], facecolor = 'indigo', label = survey.name)

    print(color)
    #plt.scatter(x,y, c=c, label = survey.name)
handles, labels = ax.get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
bbox = (1.4,0.7)
ax.legend(by_label.values(), by_label.keys(),bbox_to_anchor=bbox, fontsize=15)
fig.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
width = 0.07
yarr = np.linspace(0, 1, len(surveys))
labs = []
for i, (survey, y) in enumerate(zip(surveys, yarr)): 
    labs.append(survey.name)
    c = survey.sensitivity
    ax.fill_betweenx(np.linspace(y-width, y+width, n), survey.year[1], 2017, 
                     facecolor = survey.color, edgecolor='none', label = survey.name)
    ax.plot([], [], color=survey.color,label = survey.name)

ax.set_yticks(yarr)
ax.yaxis.tick_right()
ax.set_yticklabels(labs)
fig.autofmt_xdate()

fig.show()

fig = plt.figure()
ax = fig.add_subplot(111)
ax2 = ax.twinx()
width = 0.01
yarr = np.linspace(0, 1, len(surveys))
labs = []
colors = cm.rainbow(np.linspace(0, 1, len(surveys)))
sizes = []
for survey, y in zip(surveys, yarr): 
    labs.append(survey.name)
    s = 30*2**(survey.angular_res/30)#[20*2**n for n in range(len(x))]
    sizes.append(s)
    for freq in survey.freq:
            if len(survey.sensitivity) == 1:
                ax.scatter(freq, survey.sensitivity, s=s,
                             facecolor = survey.color, edgecolor='none', label = survey.name)
            else:
                ax.scatter(freq, survey.sensitivity[0], s=s,
                             facecolor = survey.color, edgecolor='none', label = survey.name)
ax.set_xlabel('Frequency (GHz)')
ax.set_ylabel('Sensitivity (mJY/beam)')
#ax.yaxis.set_label_position("left")
plt.xlim (-2,21)
ax2.tick_params(labeltop='off', labelright='off')
handles, labels = ax.get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
bbox = (1.4,0.93)
lgnd = ax.legend(by_label.values(), by_label.keys(),bbox_to_anchor=bbox, fontsize=15, scatterpoints=1)
for handle in lgnd.legendHandles:
    handle._sizes = [150]
    
sarr = np.array(sorted(sizes)).reshape(-1, 3).mean(axis=1)
yarr = np.arange(len(sarr))
for y,s in zip(yarr,sarr):
    ax2.scatter([],[],s=s, label=s, color = 'MediumSeaGreen')
handles2, labels2 = ax2.get_legend_handles_labels()
legendarr = sorted(list(zip([int(round(float(i),-1)) for i in labels2], handles2)), reverse=True)
leg, han = zip(*legendarr)
bbox = (1.05,-0.18)
lgnd = ax2.legend(han,leg,bbox_to_anchor=bbox, fontsize=15, scatterpoints=1, ncol = len(han), frameon=False, columnspacing=0.1)
plt.figtext(0.23,-0.1,'Angular Resolution (arcsec):', fontsize=14, ha='center')


fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 10
data = np.random.random((N, 4))
labels = ['point{0}'.format(i) for i in range(N)]

plt.subplots_adjust(bottom = 0.1)
plt.scatter(
    data[:, 0], data[:, 1], marker='o', c=data[:, 2], s=data[:, 3] * 1500,
    cmap=plt.get_cmap('Spectral'))

for label, x, y in zip(labels, data[:, 0], data[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'),
    )

plt.show()

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax2 = ax.twinx()
width = 0.01
yarr = np.linspace(0, 1, len(surveys))
labs = []
colors = cm.rainbow(np.linspace(0, 1, len(surveys)))

norm = mpl.colors.Normalize(vmin=1978, vmax=2017)
cmap = cm.Greens
m = cm.ScalarMappable(norm=norm, cmap=cmap)


sizes = []
xarr = []
zarr = []
lbs = []
colrs = []
yrs  = []
sizenames = []
for survey, y in zip(surveys, yarr): 
    labs.append(survey.name)
    s = 50*2**(survey.angular_res/30)#[20*2**n for n in range(len(x))]
    sizes.append(s)
    sizenames.append(float(survey.angular_res))
    for freq in survey.freq:
            xarr.append(freq)
            lbs.append(survey.name)
            zarr.append(survey.sensitivity[0])
            fc = m.to_rgba(survey.year[-1])
            colrs.append(fc)
            yrs.append(survey.year[-1])
            if len(survey.sensitivity) == 1:
                ax.scatter(freq, survey.sensitivity, s=s,
                             facecolor = fc, edgecolor='black', label = survey.name, alpha=0.55)
            else:
                ax.scatter(freq, survey.sensitivity[0], s=s,
                             facecolor = fc, edgecolor='black', label = survey.name, alpha=0.55)
                
ax.set_xlabel('Frequency (GHz)')
ax.set_ylabel('Sensitivity (mJY/beam)')
#ax.yaxis.set_label_position("left")
plt.xlim (-2,21)
ax2.tick_params(labeltop='off', labelright='off')
handles, labels = ax.get_legend_handles_labels()
by_label = OrderedDict(zip(labels, handles))
bbox = (1.4,0.93)
#lgnd = ax.legend(by_label.values(), by_label.keys(),bbox_to_anchor=bbox, fontsize=15, scatterpoints=1)
for handle in lgnd.legendHandles:
    handle._sizes = [150]
    
sarr = np.array(sorted(sizes)).reshape(-1, 3).mean(axis=1)
sizenames = 30 * np.log2(sarr/50)
yarr = np.arange(len(sarr))
for y,s,slabel in zip(yarr,sarr, sizenames):
    ax2.scatter([],[],s=s, label=slabel, facecolor = 'green', edgecolor = 'black',alpha =0.5)
handles2, labels2 = ax2.get_legend_handles_labels()
legendarr = sorted(list(zip([int(round(float(i),-1)) for i in labels2], handles2)), reverse=True)
leg, han = zip(*legendarr)
bbox = (1.05,-0.18)
lgnd = ax2.legend(han,['~'+str(l) for l in leg],bbox_to_anchor=bbox, fontsize=15, scatterpoints=1, ncol = len(han), frameon=False, columnspacing=0.1)
plt.figtext(0.15,-0.1,'Angular Resolution (arcsec):', fontsize=14, ha='center')


xytext=[(90, 5),(15, -30),(70, 20),(60, -40),(100, -40),(70, -25),(60, -40),(20, 20),(20, 20),(-10, 20),(50, 10),(50, 10),(50, 10),(30, 30)]
for label, x, y, xyt in zip(lbs[:-1], xarr[:-1], zarr[:-1], xytext):
    ax.annotate(
        label,
        xy=(x, y), xytext=xyt,
        textcoords='offset points', ha='right', va='bottom',
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0', alpha = 0.5, shrinkA=.9, shrinkB = .9),
        fontsize = 11
    )
ax2.axis('off')
m = cm.ScalarMappable(cmap=cmap, norm=norm)
m.set_array(yrs)
cax, kw = mpl.colorbar.make_axes(ax, location='right',fraction=0.06, pad=0.01)
cb = plt.colorbar(m, cax=cax, **kw)
cb.outline.set_visible(False)
cb.set_label('Year Completed')

fig.show()
```
